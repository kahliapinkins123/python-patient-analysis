```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Reads in Patient Data and saves to dataframe
patient = pd.read_excel("Analyst Data.xlsx")
patient.head()
```

```python
# Drops unneccessary columns 
patient = patient.drop(columns=["Eff Time","Hosp Srvc", "ZIP", "Disch Time"])
patient
```

```python
# Makes dates in the same format
patient["Eff Date"] =pd.to_datetime(patient["Eff Date"])
patient["Disch Date"] =pd.to_datetime(patient["Disch Date"])
patient.head()
```

```python
# Finds columns with missing information
patient.columns[patient.isnull().any()]
```

```python
# Fills columns with missing info as 'Unknown'
patient[['Point of Origin', 'Primary Payer', 'State', 'City',
        'Means of Arrival', 'PREF_LANGUAGE', 'INTERPRETER_REQ',
       'RACE', 'ETHNICITY', 'RELIGION', 'ORGAN_DONOR', 'AGE']] = patient[['Point of Origin', 'Primary Payer', 'State', 'City',
        'Means of Arrival', 'PREF_LANGUAGE', 'INTERPRETER_REQ',
       'RACE', 'ETHNICITY', 'RELIGION', 'ORGAN_DONOR', 'AGE']].fillna("Unknown")
```

```python
patient.head()
```

```python
patient["ORGAN_DONOR"].unique()
```

```python
patient["ORGAN_DONOR"].unique()
```

```python
# Fills columns with missing data (under a different name) as unknown
patient["ORGAN_DONOR"] = np.where(patient["ORGAN_DONOR"].isin(['Does not wish to discuss', 'Patient not available']), 'Unknown', patient["ORGAN_DONOR"])
patient.head()
```

```python
# Creates column with more general data
patient["general_means_of_arrival"] = np.where(patient["Means of Arrival"].str.contains('ambulance'), 'ambulance', patient["Means of Arrival"])
patient.head()
```

```python
# Fills columns with missing data (under a different name) as unknown
patient["RACE"] = np.where(patient["RACE"].isin(['Patient Refused']), 'Unknown', patient["RACE"])
patient["RACE"].fillna('Unknown', inplace=True)
patient.head()
```

```python
# Creates age buckets: 0-9, 10-19, 20-29, 30-39,40-49, 50-59, 60-69, 70-79, 80+

def assign_age_buckets(age):
    if age in range(0, 10):
        return "0-9"
    elif age in range(10, 20):
        return "10-19"
    elif age in range(20, 30):
        return "20-29"
    elif age in range(30, 40):
        return "30-39"
    elif age in range(40, 50):
        return "40-49"
    elif age in range(50, 60):
        return "50-59"
    elif age in range(60, 70):
        return "60-69"
    elif age in range(70, 80):
        return "70-79"
    elif age in range(80, 110):
        return "80+"
    else:
        "Unknown"
```

```python
# Creates column with age ranges
patient["age_range"] = patient["AGE"].apply(assign_age_buckets)
```

```python
patient["ETHNICITY"].unique()
```

```python
patient["ETHNICITY"].unique()
```

```python
# # Fills columns with missing data (under a different name) as unknown, and shortens data names
patient["ETHNICITY"] = np.where(patient["ETHNICITY"].isin(['Patient Refused', 'Hispanic/Latino ? unspecified']), 'Unknown', patient["ETHNICITY"])
patient["ETHNICITY"] = np.where(patient["ETHNICITY"].isin(['Mexican, Mexican American, Chicano/a']), 'Mexican', patient["ETHNICITY"])
patient["ETHNICITY"] = np.where(patient["ETHNICITY"].isin(['Not of Spanish/Hispanic Origin']), 'Not Hispanic', patient["ETHNICITY"])

patient.head()
```

```python
def is_hispanic(hispanic):
    if "Not Hispanic" in hispanic:
        return "N"
    elif hispanic == 'Unknown':
        return "Unknown"
    else:
        return "Y"
```

```python
# Creates column for hispanic vs nonhispanic
patient['is_hispanic'] = patient["ETHNICITY"].apply(is_hispanic)
patient.head()
```

```python
# Creates uniform column names
patient.columns = ['eff_date', 'location', 'sex', 'origin', 'primary_payer',
       'second_payer', 'state', 'city', 'disch_date', 'means_of_arrival',
       'pref_language', 'interpreter_req', 'race', 'race_2', 'ethnicity',
       'religion', 'organ_donar', 'age', 'general_means_of_arrival',
       'age_range', 'is_hispanic']
```

```python
patient.head()
```

```python
# Creates a column for counting data
patient["count"] = 1
```

```python
patient.race.unique()
```

```python
def simplify_race(race):
    if race == 'African American (Black)':
        return 'African American'
    elif race == 'Other Race':
        return 'Other'
    elif race == 'Native American (American Indian/Eskimo/Aleutian)':
        return 'Native American'
    elif race in ['Filipino','Asian - unspecified', 'Chinese', 'Hmong', 'Japanese', 'Korean', 'Cambodian','Thai', 'Vietnamese', 'Laotian', 'Indonesian']:
        return 'Asian'
    elif race in ['Sri Lankan', 'Samoan','Other Pacific Islander', 'Guamanian or Chamorro', 'Native Hawaiian', 'Native Hawaiian or Other Pacific Islander']:
        return 'Pacific Islander'
    elif race in ['Pakistani', 'Bangladeshi']:
        return 'Middle Eastern'
    elif race == 'Asian Indian':
        return 'Indian'
    else:
        return race
```

```python
# More general race definitions
patient["race"] = patient["race"].apply(simplify_race)
```

```python
# Makes missing values Unknown
patient["origin"].unique()
patient["origin"] = np.where(patient["origin"] == 'information unavailable', 'Unknown', patient["origin"])
```

```python
patient["religion"].unique()
patient["religion"] = np.where(patient["religion"] == 'Patient Refused', 'Unknown', patient["religion"])
```
